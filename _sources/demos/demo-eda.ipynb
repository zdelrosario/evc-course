# Setup

```python
# 1. Main package import
import grama as gr  
# 2. Set up grama
DF = gr.Intention() 
# 3. Show figures correctly
%matplotlib inline
```

Load a dataset


```python
from grama.data import df_diamonds
```

# Simple checks


```python
(
    df_diamonds
    >> gr.tf_head()
)
```

```python
(
    df_diamonds
    >> gr.tf_describe()
)
```

# Visualize


## 1d EDA


### Carat


```python
(
    df_diamonds
    >> gr.ggplot(gr.aes("carat"))
    + gr.geom_histogram()
)
```

Focus on a range of `carat` values


```python
(
    df_diamonds
    >> gr.tf_filter(DF.carat <= 1.0)
    >> gr.ggplot(gr.aes("carat"))
    + gr.geom_histogram()
)
```

Refine the bin count to better show patterns


```python
(
    df_diamonds
    >> gr.tf_filter(DF.carat <= 1.0)
    >> gr.ggplot(gr.aes("carat"))
    + gr.geom_histogram(bins=80)
)
```

## 2d EDA


### Price v Carat


```python
(
    df_diamonds
    >> gr.ggplot(gr.aes("carat", "price"))
    + gr.geom_point()
)
```

Modify the dots to illustrate patterns


```python
(
    df_diamonds
    >> gr.ggplot(gr.aes("carat", "price"))
    + gr.geom_point(size=1/8, alpha=1/5)
)
```
