## Preview: Using layers


```python
df_diamonds

(
    df_diamonds
    
    >> gr.ggplot(gr.aes("carat", "price"))
    + gr.geom_point(
        data=df_diamonds
        >> gr.tf_filter(DF.cut == "Fair"),
        color="black",
    )
    + gr.geom_point(gr.aes(color="cut"), size=0.5)
)
```
